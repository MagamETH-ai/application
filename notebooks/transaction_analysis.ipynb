```python
import pandas as pd
import os
```

```python
dfs = []
for f in os.listdir('/content/'):
    if f.startswith('export'):
        dfs.append(pd.read_csv(f))
```

```python
dfs[0]
```

```python
data = pd.concat(dfs)
```

```python
popular_wallets = data[['To', 'Method']].value_counts()
popular_wallets
```

```python
for row in popular_wallets.items():
    print(row[0][0], row[0][1], row[1])
```
